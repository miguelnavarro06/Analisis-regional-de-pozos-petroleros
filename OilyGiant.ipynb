# 2. Entrenamiento y prueba del modelo para cada región

```python
def train_and_evaluate(df):
    features = df.drop('product', axis=1)
    target = df['product']
    
    # Dividir los datos 
    features_train, features_valid, target_train, target_valid = train_test_split(
        features, target, test_size=0.25, random_state=12345)
    
    # Entrenar modelo
    model = LinearRegression()
    model.fit(features_train, target_train)
    
    # Predicciones
    predictions = model.predict(features_valid)
    predictions = pd.Series(predictions, index=target_valid.index)

    #Calculo del error
    rmse = mean_squared_error(target_valid, predictions) ** 0.5
    avg_predicted = predictions.mean()
    
    return target_valid, predictions, rmse, avg_predicted

# Ejecutar para las 3 regiones

results = {}
for i, df in enumerate([data_0, data_1, data_2]):
    target_valid, pred, rmse, avg = train_and_evaluate(df)
    results[i] = {'target': target_valid, 'predictions': pred}
    print(f"Región {i}: RMSE = {rmse:.2f}, Promedio predicho = {avg:.2f}")
```

# 3. Cálculo de ganancias

```python
#Definir variables 

presupuesto = 100_000_000
pozos_a_abrir = 200
ingresos_por_unidad = 4500

# Cálculo del volumen mínimo necesario por pozo
costo_por_pozo= presupuesto / pozos_a_abrir
unidades_necesarias = costo_por_pozo / ingresos_por_unidad

print(f"Volumen mínimo necesario por pozo: {unidades_necesarias:.2f} mil barriles")
```

Al comparar este valor con el promedio de reservas de las regiones (Región 0: 92.5, Región 1: 68.7, Región 2: 95), se observa que ninguna región alcanza el punto de equilibrio de forma natural.

# 4. Cálculo de ganancia 

```python
#Definición de la funcion para el cálculo

def calculate_profit(target, predictions, count):
    
    # Seleccionar los índices de las predicciones más altas
    preds_sorted = predictions.sort_values(ascending=False)
    selected_targets = target[preds_sorted.index][:count]
    
    total_volume = selected_targets.sum()
    revenue = total_volume * ingresos_por_unidad
    return revenue - presupuesto
```

```python
#Predicciones para los pozos para las 3 regiones

for i in range(3):
    target = results[i]['target']
    predictions = results[i]['predictions']
    
    # Cálculola ganancia
    ganancia = calculate_profit(target, predictions, pozos_a_abrir)
    top_preds = predictions.sort_values(ascending=False)[:pozos_a_abrir]
    volumen_real_top = target[top_preds.index].sum()
    
    print(f"--- REGIÓN {i} ---")
    print(f"Volumen total real de los 200 mejores pozos: {volumen_real_top:.2f} mil barriles")
    print(f"Ganancia potencial: ${ganancia:,.2f} USD\n")
```

Según el beneficio máximo obtenido en el cálculo, la región recomendada para el desarrollo es la región 0, debido a que presenta el mayor margen de beneficio. Aunque esta región destaca por su rentabilidad potencial, sin embargo, dicha región presenta un RMSE elevado, lo que implica que la incertidumbre en las prediccion es significativa.

# 5. Cálculo de riesgo y ganancia para cada región

```python
import numpy as np

state = np.random.RandomState(12345)

for i in range(3):
    target = results[i]['target']
    predictions = results[i]['predictions']
    values = []
    
    for _ in range(1000):
        # Seleccionamos 500 puntos al azar (estudio geológico)
        target_subsample = target.sample(n=500, replace=True, random_state=state)
        probs_subsample = predictions[target_subsample.index]
        
        # Calculamos ganancia de los mejores 200 de esos 500
        profit = calculate_profit(target_subsample, probs_subsample, pozos_a_abrir)
        values.append(profit)
    
    values = pd.Series(values)
    mean_profit = values.mean()
    risk = (values < 0).mean() * 100
    lower = values.quantile(0.025)
    upper = values.quantile(0.975)
    
    print(f"--- REGIÓN {i} ---")
    print(f"Beneficio promedio: {mean_profit:,.2f} USD")
    print(f"Intervalo de confianza (95%): ({lower:,.2f}, {upper:,.2f})")
    print(f"Riesgo de pérdida: {risk:.2f}%")
```
